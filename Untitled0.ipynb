```python
import pandas as pd

data = {
    "Name": ["Amit", "Priya", "Rahul", "Sneha"],
    "Maths": [85, 92, 70, 88],
    "Science": [78, 88, 65, 90],
    "English": [90, 95, 72, 85]
}

df = pd.DataFrame(data)

df.to_csv("students.csv", index=False)

print("CSV file created!")
```

```python
import pandas as pd

data = pd.read_csv("students.csv")

print("\nStudent Data:\n")
print(data)

data["Total"] = data["Maths"] + data["Science"] + data["English"]
data["Average"] = data["Total"] / 3

topper = data.loc[data["Total"].idxmax()]

print("\nUpdated Data:\n")
print(data)

print("\nTopper:\n")
print(topper)
```
